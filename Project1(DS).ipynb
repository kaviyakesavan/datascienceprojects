```python
# @title PROJECT 1
# analysing IMDB Movie dataset
import pandas as pd
df = pd.read_csv("/content/IMDBMovie (1).csv")
df
```

```python
# @title Task 1
# how many rows are there in the IMDB dataset
df.shape[0]
```

```python
# @title Task 2
# what ist the 75th percentile of rating in the imdb dataset?
df.describe()
```

```python
# @title Task 3
#how many NA values are there in the field "Revenue"?
df["Revenue_millions"].isnull().sum()
```

```python
# @title Task 4
#how many movies have  revenue higher than 75 million?
(df['Revenue_millions'] > 75).sum()
df[df["Revenue_millions"]>75]["Revenue_millions"].count()
```

```python
# @title Task 5
# How many movies have revenue greater than 50 million but rating less than 7?
df[(df['Revenue_millions'] > 50) & (df['Rating'] < 7)]['Title'].count()
```

```python
# @title Task 6
# what is the total Revenue genarated by movies in the year 2015 ?
df[df["Year"] == 2015]["Revenue_millions"].sum()
```

```python
# @title Task 7
# What is the average rating for the genre adventure in the year 2015?
df[(df["Year"] == 2015) & (df["Genre"] == "Adventure")]["Rating"].mean()
```

```python
# @title Task 8
# What is the average duration of movies in rows 75 to 150? Please note that the rows in python start from 0.
df.loc[75:150]["Runtime_minutes"].mean()
```

```python
# @title Task 9
# Which year generated the highest revenue?
df['Revenue_millions'].max()
df[df['Revenue_millions']== 936.63 ]["Year"]
df.sort_values(by = "Revenue_millions", ascending = False)
```

```python
# @title Task 10
# What is the maximum revenue out of (10,20,30,40,50) rows]
df.drop([10,20,30,40,50],axis =0)["Revenue_millions"].max()
```

```python
# @title Task 11
# How many movies with the genres â€˜Adventureâ€™, â€˜Actionâ€™, â€˜Horrorâ€™, and â€˜Crimeâ€™ exist in the IMDB dataset?
cond = df[df['Genre'].isin(['Adventure','Action','Horror','Crime'])]['Title'].count()
cond
```

```python
# @title Task 12
# Create a genre-level report with metrics average rating, the average number of votes, and average revenue.
# What is the average rating of the â€˜Horrorâ€™ genre?
df[df["Genre"] == "Horror"]["Rating"].mean()
```

```python
# @title Task 13
# How many movies has Billy Ray	directed and find the year of release of those movies
df[df["Director"] == 'Billy Ray'].sum()
```

```python
# @title Task 14
# How many movies were released in the year 2012 - 14. What type of genre were released the most
cond = df[df['Year'].isin([2012,2013,2014])]['Year'].count()
print("No.Of Movies released in the year 2012 - 14 were",cond)
cond1 = df[df['Year'].isin([2012,2013,2014])]['Genre']
print("Type of genre were released the most",cond1.max())
```

```python
# @title Task 15
# Which Movie had the highest vote and what genre it belongs to.
print(df[df['Votes']== df["Votes"].max()]['Title'])
df[df['Votes']== df["Votes"].max()]['Genre']
```

```python
# @title Task 16
# Find the Director whose movie grossed the highest. Also find the total revenue generated by that director
cond = df[df['Revenue_millions'] == df['Revenue_millions'].max()]['Director']
print("Director whose movie grossed the highest is",cond)
cond1 = df[df["Director"]== 'J.J. Abrams']["Revenue_millions"].sum()
print("Total Revenue generated by JJ Abrams", cond1)
```
